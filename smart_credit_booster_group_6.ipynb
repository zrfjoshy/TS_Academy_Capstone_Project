## Project Description

This project develops a machine learning–based credit scoring system that classifies individuals into credit score categories — `Good`, `Standard`, or `Poor` — using financial and behavioural attributes. Traditional credit bureaus rely on formal banking histories that exclude large portions of the population in emerging markets. By leveraging publicly available financial behaviour data and supervised classification algorithms, this project demonstrates how data science can produce fair, interpretable, and scalable credit assessments applicable to underserved populations, with particular relevance to the Nigerian fintech ecosystem.

---
## 1. Project Overview

Access to credit is one of the most critical enablers of economic mobility. In developing economies, a significant proportion of the working population — including informal traders, contract workers, and students — lacks the formal banking history required by conventional credit scoring models. This phenomenon, commonly referred to as being *credit invisible*, effectively excludes financially responsible individuals from accessing loans, mortgages, and other financial instruments.

Traditional credit scores such as the FICO model are computed from credit bureau records: loan repayment history, credit utilisation, and length of credit history. While reliable for banked individuals, these metrics fail entirely for those without documented financial footprints. The result is a systemic exclusion of lower-income populations from the formal economy — a challenge particularly acute in sub-Saharan Africa, where the majority of economic activity occurs in informal sectors.

Machine learning offers a promising alternative by identifying predictive patterns in broader financial behaviour data — payment consistency, debt-to-income ratios, number of active accounts, and spending habits — without requiring a prior credit bureau record. This project applies supervised classification techniques to construct a model capable of distinguishing between `Good`, `Standard`, and `Poor` credit profiles, with an emphasis on model interpretability, fairness, and reproducibility.

---
## 2. Problem Statement

Conventional credit scoring systems present several well-documented limitations. First, they depend on historical credit data, which is unavailable for a large proportion of the global population. Second, they tend to perpetuate socioeconomic bias, as individuals from lower-income backgrounds or informal employment sectors are systematically underscored. Third, traditional models offer little transparency — a rejected applicant is rarely told which factor triggered the decision or how to improve their standing. Fourth, these systems are slow and expensive to deploy in emerging markets where digital infrastructure is still developing.

There is therefore a clear need for a fast, fair, and explainable credit assessment system that leverages alternative digital and financial signals to evaluate creditworthiness without requiring a prior formal credit history.

This project addresses these limitations by framing creditworthiness prediction as a **multi-class classification problem**. Given a set of financial and behavioural features, the objective is to assign each individual to one of three credit tiers: `Good`, `Standard`, or `Poor`. Classification models are appropriate here because the outcome is discrete and ordinal, the dataset contains labelled examples suitable for supervised learning, and the resulting model can be evaluated and explained using well-established metrics.

## 7. Data Cleaning and Preprocessing

The dataset contains several well-documented quality issues that must be addressed before any analysis can take place. These include: non-standard null representations (`_`, `NA`, `#F%$@D!`, `!@9#%8`), numeric columns contaminated with trailing underscores, impossible values in `Age`, a free-text `Credit_History_Age` field requiring regex parsing, and a multi-value `Type_of_Loan` column. Each step below is applied deliberately and documented with justification.

```python
df = df_raw.copy()

# Drop identifier and PII columns 
drop_cols = ['ID', 'Customer_ID', 'Name', 'SSN', 'Month']
df.drop(columns=drop_cols, inplace=True)
print(f'After dropping identifiers: {df.shape}')
```

Identifier and PII columns (`ID`, `Customer_ID`, `Name`, `SSN`) carry no predictive information and must be removed to prevent data leakage. `Month` is excluded as the model treats each row as an independent financial snapshot rather than a temporal sequence.

```python
# Standardise all non-standard null markers to np.nan
NULL_MARKERS = ['_', 'NA', 'nan', 'NaN', '', '#F%$@D!', '!@9#%8', 'null', 'NULL']
df.replace(NULL_MARKERS, np.nan, inplace=True)
print('Null markers standardised to np.nan')
```

The raw dataset uses multiple non-standard representations for missing values. Unifying these to `np.nan` before any type casting prevents silent conversion errors that would corrupt numeric columns.

```python
# Robust numeric cleaning
def clean_numeric(series):
    return (
        series.astype(str)
              .str.replace(r'[^0-9.\-]', '', regex=True)
              .replace('', np.nan)
    )

numeric_cols = [
    'Age', 'Annual_Income', 'Monthly_Inhand_Salary', 'Num_Bank_Accounts',
    'Num_Credit_Card', 'Interest_Rate', 'Num_of_Loan', 'Delay_from_due_date',
    'Num_of_Delayed_Payment', 'Changed_Credit_Limit', 'Num_Credit_Inquiries',
    'Outstanding_Debt', 'Credit_Utilization_Ratio', 'Total_EMI_per_month',
    'Amount_invested_monthly', 'Monthly_Balance'
]

for col in numeric_cols:
    df[col] = pd.to_numeric(clean_numeric(df[col]), errors='coerce')

print('Numeric columns cleaned and cast.')
```

Several numeric columns contain trailing underscores (e.g. `28_`, `4_`) and other non-numeric characters introduced during data collection. The `clean_numeric` function strips all non-numeric characters using regex before coercing to float, with `errors='coerce'` converting any remaining non-numeric values to `NaN` rather than raising an error.

```python
# Fix impossible Age values
df.loc[(df['Age'] < 18) | (df['Age'] > 100), 'Age'] = np.nan
print(f'Invalid Age values replaced with NaN.')
```

The raw `Age` column contains values such as `-500`, which are physiologically impossible. All ages outside the range 18–100 are nullified and will be imputed using the column median in a later step.

```python
# Step 5: Parse Credit_History_Age from string to integer months
def parse_credit_history(val):
    if pd.isna(val):
        return np.nan
    years  = re.search(r'(\d+)\s*Year',  str(val))
    months = re.search(r'(\d+)\s*Month', str(val))
    y = int(years.group(1))  if years  else 0
    m = int(months.group(1)) if months else 0
    return y * 12 + m

df['Credit_History_Months'] = df['Credit_History_Age'].apply(parse_credit_history)
df.drop(columns=['Credit_History_Age'], inplace=True)
print('Credit_History_Age parsed to integer months as Credit_History_Months.')
```

`Credit_History_Age` is stored as a free-text string (e.g. `"22 Years and 3 Months"`). A regex parser extracts the year and month components and converts them to a single integer representing total months of credit history. This transformation makes the feature directly usable by machine learning models.

```python
# Ordinal encode Credit_Mix (Bad < Standard < Good)
credit_mix_map = {'Bad': 0, 'Standard': 1, 'Good': 2}
df['Credit_Mix'] = df['Credit_Mix'].map(credit_mix_map)

# Binary encode Payment_of_Min_Amount
df['Payment_of_Min_Amount'] = df['Payment_of_Min_Amount'].map({'Yes': 1, 'No': 0, 'NM': np.nan})

# Remove garbage entries from Payment_Behaviour then one-hot encode
valid_pb = df['Payment_Behaviour'].value_counts()
valid_pb = valid_pb[valid_pb > 100].index.tolist()
df.loc[~df['Payment_Behaviour'].isin(valid_pb), 'Payment_Behaviour'] = np.nan
df = pd.get_dummies(df, columns=['Payment_Behaviour'], prefix='PB', dummy_na=False)

# Derive loan type count from multi-label column
def count_loan_types(val):
    if pd.isna(val): return 0
    return len([x for x in str(val).split(',') if x.strip()])

df['Num_Loan_Types'] = df['Type_of_Loan'].apply(count_loan_types)
df.drop(columns=['Type_of_Loan'], inplace=True)

# One-hot encode Occupation
df = pd.get_dummies(df, columns=['Occupation'], prefix='Occ', dummy_na=False)

print(f'Encoding complete. Shape: {df.shape}')
```

`Credit_Mix` is ordinally encoded because it reflects a natural quality ordering (Bad → Standard → Good). `Payment_Behaviour` contains garbage entries (e.g. `!@9#%8`) which are nullified before one-hot encoding. `Type_of_Loan` is a multi-label field that is converted to a count of distinct loan types, preserving its numeric signal without requiring multi-hot encoding. `Occupation` is one-hot encoded as it is a nominal categorical variable with no inherent order.

```python
# Median imputation for all remaining numeric nulls
num_features_all = df.select_dtypes(include=[np.number]).columns.tolist()
if 'Credit_Score_Enc' in num_features_all:
    num_features_all.remove('Credit_Score_Enc')

imputer = SimpleImputer(strategy='median')
df[num_features_all] = imputer.fit_transform(df[num_features_all])

remaining_nulls = df.isnull().sum().sum()
print(f'Total remaining nulls after imputation: {remaining_nulls}')
```

Median imputation is chosen for all remaining numerical missing values because the median is robust to the outliers and right-skewed distributions observed in financial data. Mean imputation would be distorted by extreme values in columns such as `Annual_Income` and `Outstanding_Debt`. The imputer is fitted on the full preprocessed dataset before the train-test split in the modelling section.

The bar chart quantifies missingness per feature, while the heatmap provides a row-level perspective. This supports median/mode imputation as the appropriate handling strategy rather than row deletion, which would unnecessarily reduce the dataset size.

## 9. Feature Engineering

Three derived features are constructed to capture financial stress signals that are present in the raw data but not directly represented as single columns. All engineered features are derived exclusively from existing dataset columns — no synthetic or external data is introduced.

```python
# Debt_to_Income_Ratio: proportion of annual income consumed by outstanding debt
df['Debt_to_Income_Ratio'] = df['Outstanding_Debt'] / (df['Annual_Income'] + 1)

# EMI_Burden: fraction of monthly salary consumed by EMI obligations  
df['EMI_Burden'] = df['Total_EMI_per_month'] / (df['Monthly_Inhand_Salary'] + 1)

# Credit_Pressure_Index: compound delinquency signal (delay magnitude × frequency)
df['Credit_Pressure_Index'] = df['Delay_from_due_date'] * df['Num_of_Delayed_Payment']

print('Feature Engineering Summary:')
print(df[['Debt_to_Income_Ratio','EMI_Burden','Credit_Pressure_Index']].describe().round(4).to_string())
```

`Debt_to_Income_Ratio` is a standard credit underwriting metric that measures how heavily a customer's debt burden consumes their income. High values indicate financial overextension and are associated with elevated default risk. A constant of 1 is added to the denominator to prevent division by zero.

`EMI_Burden` captures the proportion of take-home pay committed to loan repayments. Values approaching or exceeding 1.0 indicate that a customer's repayment obligations consume most or all of their salary — a strong indicator of financial stress.

`Credit_Pressure_Index` combines payment delay magnitude (average days overdue) with payment delay frequency, producing a single compound delinquency severity score. This captures customers who are both frequently and severely late, which is more informative than either metric alone.

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 4))
eng_feats = ['Debt_to_Income_Ratio', 'EMI_Burden', 'Credit_Pressure_Index']

for i, feat in enumerate(eng_feats):
    for label, color in class_colors.items():
        mask = df_raw['Credit_Score'] == label
        axes[i].hist(df[mask][feat].clip(upper=df[feat].quantile(0.99)),
                     bins=30, alpha=0.5, label=label, color=color, edgecolor='white')
    axes[i].set_title(feat, fontweight='bold')
    axes[i].set_xlabel('Value')
    axes[i].legend(fontsize=8)

plt.suptitle('Engineered Feature Distributions by Credit Score Class', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig("Engineered_Feature_Distributions_Stratified_by_Credit_Score_Class.png", dpi=300)
plt.show()
```

All three engineered features demonstrate class-level separation, confirming that they capture meaningful discriminatory signal. `Debt_to_Income_Ratio` and `Credit_Pressure_Index` particularly show strong separation between `Good` and `Poor` classes, validating their inclusion in the modelling phase.

## 10. Data Preparation for Modelling

```python
# Encode target variable
le = LabelEncoder()
y = le.fit_transform(df_raw['Credit_Score'])
print('Class encoding:', dict(zip(le.classes_, le.transform(le.classes_))))

X = df.select_dtypes(include=[np.number])
print(f'Feature matrix shape: {X.shape}')
```

```python
# Final imputation pass
final_imputer = SimpleImputer(strategy='median')
X = pd.DataFrame(final_imputer.fit_transform(X), columns=X.columns)

# Stratified 80/20 train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE, stratify=y
)
print(f'Training set:  {X_train.shape}')
print(f'Test set:      {X_test.shape}')
print()
print('Class distribution — Training set:')
pd.Series(y_train).value_counts().rename(index=dict(zip(le.transform(le.classes_), le.classes_)))
```

```python
# Feature scaling for Logistic Regression
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)
```

A stratified 80/20 train-test split preserves the original class proportions in both subsets — essential given the class imbalance identified in Section 8.3. Feature scaling (StandardScaler) is applied only for Logistic Regression, as tree-based models are scale-invariant. The scaler is fitted exclusively on the training set and applied to the test set to prevent data leakage. All preprocessing steps respect the training/test boundary.

## 11. Handling Class Imbalance

```python
print('Class distribution before SMOTE:')
before = pd.Series(y_train).value_counts().rename(
    index=dict(zip(le.transform(le.classes_), le.classes_)))
print(before)

smote = SMOTE(random_state=RANDOM_STATE)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

print('\nClass distribution after SMOTE:')
after = pd.Series(y_train_res).value_counts().rename(
    index=dict(zip(le.transform(le.classes_), le.classes_)))
print(after)

# Rescale SMOTE data for Logistic Regression
X_train_res_scaled = scaler.fit_transform(X_train_res)
X_test_scaled = scaler.transform(X_test)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
colors_bar = ['#2ecc71','#e74c3c','#f39c12']

before.plot(kind='bar', ax=axes[0], color=colors_bar, edgecolor='white', rot=0)
axes[0].set_title('Before SMOTE', fontweight='bold')
axes[0].set_ylabel('Count')
axes[0].set_ylim(0, before.max() * 1.15)
for i, v in enumerate(before.values):
    axes[0].text(i, v + 200, f'{v:,}', ha='center', fontsize=9)

after.plot(kind='bar', ax=axes[1], color=colors_bar, edgecolor='white', rot=0)
axes[1].set_title('After SMOTE', fontweight='bold')
axes[1].set_ylabel('Count')
axes[1].set_ylim(0, after.max() * 1.15)
for i, v in enumerate(after.values):
    axes[1].text(i, v + 200, f'{v:,}', ha='center', fontsize=9)

plt.suptitle('Class Distribution Before and After SMOTE', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig("Class_Distribution_of_Smote.png", dpi=300)
plt.show()
```

SMOTE (Synthetic Minority Over-sampling Technique) generates synthetic samples for minority classes by interpolating between existing observations in feature space, rather than simply duplicating rows. This introduces diversity in the minority class representation and avoids the overfitting risk of random duplication. After SMOTE, all three classes are equally represented in the training data, ensuring that models learn decision boundaries for each class with equal emphasis. Critically, SMOTE is applied only to the training set — the test set remains untouched to ensure evaluation reflects real-world class distributions.

## Evaluation Metrics — Justification

The following metrics are used to evaluate all classification models, in accordance with the capstone classification track requirements:

| Metric | Purpose |
|---|---|
| **Accuracy** | Overall proportion of correct predictions across all classes |
| **Precision** | Of all predicted positives, what fraction were truly positive — measures false alarm rate |
| **Recall** | Of all actual positives, what fraction were correctly identified — measures miss rate |
| **F1-Score** | Harmonic mean of precision and recall — balances both, especially valuable under class imbalance |
| **Confusion Matrix** | Full breakdown of correct and incorrect predictions per class — enables FP/FN error analysis |

Accuracy alone is insufficient for imbalanced datasets, as a model predicting only the majority class can still achieve high accuracy. The weighted F1-score is therefore the primary performance metric for model comparison in this project, as it accounts for class frequency when averaging across classes.

> ⚠️ Note: RMSE and other regression metrics are not used anywhere in this project. This project is a classification task and uses only classification-appropriate metrics.

## 13. Advanced Model — Random Forest

**Model Type:** Advanced  
To capture non-linear relationships and feature interactions present in credit behaviour data, two ensemble models are implemented. Random Forest is the first advanced model, leveraging bagged decision trees to reduce variance and model complex patterns.

```python
rf = RandomForestClassifier(
    n_estimators=200, max_depth=15,
    min_samples_leaf=5, random_state=RANDOM_STATE, n_jobs=-1
)
rf.fit(X_train_res, y_train_res)
y_pred_rf = rf.predict(X_test)

print('=== Random Forest — Classification Report ===')
print(classification_report(y_test, y_pred_rf, target_names=le.classes_))
```

```python
cm_rf = confusion_matrix(y_test, y_pred_rf)
fig, ax = plt.subplots(figsize=(6, 5))
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Blues',
            xticklabels=le.classes_, yticklabels=le.classes_, ax=ax, linewidths=0.5)
ax.set_xlabel('Predicted Label', fontsize=11)
ax.set_ylabel('True Label', fontsize=11)
ax.set_title('Confusion Matrix — Random Forest', fontweight='bold')
plt.tight_layout()
plt.savefig("Confusion_Matrix_rf.png", dpi=300)
plt.show()
```

Random Forest is an ensemble of independently trained decision trees, each trained on a bootstrapped sample with a random subset of features considered at each split. This design reduces variance compared to a single decision tree while retaining the ability to model complex non-linear interactions. The depth constraint (`max_depth=15`) and minimum leaf size (`min_samples_leaf=5`) are regularisation choices that prevent individual trees from overfitting to noise. The performance improvement over Logistic Regression is expected given the non-linear nature of credit risk relationships.

## 14. Advanced Model — XGBoost

**Model Type:** Advanced  
XGBoost is the second advanced model, using gradient boosting to sequentially correct prediction errors. It typically outperforms Random Forest on structured tabular data due to its sequential error-correction mechanism.

```python
xgb = XGBClassifier(
    n_estimators=300, max_depth=6, learning_rate=0.1,
    subsample=0.8, colsample_bytree=0.8,
    use_label_encoder=False, eval_metric='mlogloss',
    random_state=RANDOM_STATE, n_jobs=-1
)
xgb.fit(X_train_res, y_train_res)
y_pred_xgb = xgb.predict(X_test)

print('=== XGBoost — Classification Report ===')
print(classification_report(y_test, y_pred_xgb, target_names=le.classes_))
```

```python
cm_xgb = confusion_matrix(y_test, y_pred_xgb)
fig, ax = plt.subplots(figsize=(6, 5))
sns.heatmap(cm_xgb, annot=True, fmt='d', cmap='Blues',
            xticklabels=le.classes_, yticklabels=le.classes_, ax=ax, linewidths=0.5)
ax.set_xlabel('Predicted Label', fontsize=11)
ax.set_ylabel('True Label', fontsize=11)
ax.set_title('Confusion Matrix — XGBoost', fontweight='bold')
plt.tight_layout()
plt.savefig("Confusion_Matrix_Xgb.png", dpi=300)
plt.show()
```

XGBoost implements gradient-boosted decision trees using a sequential learning approach: each successive tree is trained to correct the residual errors of the previous ensemble. The `subsample` and `colsample_bytree` parameters introduce stochasticity into the boosting process to prevent overfitting. The conservative `learning_rate` of 0.1 allows gradual, stable learning across 300 boosting rounds. XGBoost's sequential error-correction mechanism typically yields stronger performance than Random Forest's parallel averaging approach on structured tabular data.

## 15. Model Comparison

```python
def get_metrics(y_true, y_pred, name):
    return {
        'Model':     name,
        'Accuracy':  round(accuracy_score(y_true, y_pred), 4),
        'Precision': round(precision_score(y_true, y_pred, average='weighted'), 4),
        'Recall':    round(recall_score(y_true, y_pred, average='weighted'), 4),
        'F1 Score':  round(f1_score(y_true, y_pred, average='weighted'), 4),
    }

results = pd.DataFrame([
    get_metrics(y_test, y_pred_lr,  'Logistic Regression'),
    get_metrics(y_test, y_pred_rf,  'Random Forest'),
    get_metrics(y_test, y_pred_xgb, 'XGBoost'),
]).set_index('Model')

print(results.to_string())
```

The comparison table quantifies the performance gap between the baseline and ensemble models across all required metrics. Logistic Regression underperforms due to its linear decision boundary assumption, which is inappropriate for the interaction-driven nature of credit risk. Random Forest delivers a substantial improvement by modelling non-linear feature relationships. XGBoost is expected to deliver the strongest overall performance, particularly in weighted F1 score which accounts for class imbalance in the evaluation. The weighted averaging of precision, recall, and F1 is appropriate here given the class imbalance present in the test set.

## 16. Confusion Matrix Interpretation — Error Analysis

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
model_results = [
    ('Logistic Regression', y_pred_lr),
    ('Random Forest',       y_pred_rf),
    ('XGBoost',             y_pred_xgb)
]

for ax, (name, preds) in zip(axes, model_results):
    cm = confusion_matrix(y_test, preds)
    cm_pct = cm.astype(float) / cm.sum(axis=1, keepdims=True)
    sns.heatmap(cm_pct, annot=True, fmt='.2%', cmap='Blues',
                xticklabels=le.classes_, yticklabels=le.classes_,
                ax=ax, linewidths=0.5)
    ax.set_title(name, fontweight='bold')
    ax.set_xlabel('Predicted')
    ax.set_ylabel('Actual')

plt.suptitle('Row-Normalised Confusion Matrices (% of Actual Class)', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Detailed FP/FN analysis for best model (XGBoost)
cm_xgb_detail = confusion_matrix(y_test, y_pred_xgb)
classes = le.classes_

print('=== XGBoost Error Analysis ===')
for i, actual in enumerate(classes):
    for j, predicted in enumerate(classes):
        if i != j:
            count = cm_xgb_detail[i, j]
            pct = count / cm_xgb_detail[i].sum() * 100
            print(f'  Actual: {actual:8s} → Predicted: {predicted:8s} | Count: {count:5,} ({pct:.1f}%)')
```

Row-normalised confusion matrices reveal the percentage of actual class observations that are correctly and incorrectly classified by each model.

In the context of credit scoring, two error types carry distinct real-world consequences:

**False Positives (predicting `Good` when actual is `Poor`):** This is the most costly misclassification for a lender. A financial institution acting on this prediction would extend credit to a high-risk customer, resulting in potential loan defaults and financial losses. In the interest of financial stability, minimising this error type is critical.

**False Negatives (predicting `Poor` when actual is `Good`):** These represent missed lending opportunities. Creditworthy individuals are incorrectly denied access to credit products, reinforcing the financial exclusion problem this project aims to address. In a financial inclusion context, this error type is particularly harmful.

The ensemble models, particularly XGBoost, substantially reduce both error types compared to Logistic Regression, making them more suitable for deployment in a credit decision support system. The most dangerous boundary — misclassifying `Poor` as `Good` — is minimised most effectively by XGBoost's boosting mechanism, which focuses corrective learning on the most frequently misclassified observations.

## 18. ROC–AUC Curve Analysis

```python
y_test_bin = label_binarize(y_test, classes=[0, 1, 2])
roc_colors = ['#2ecc71', '#f39c12', '#e74c3c']

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

for ax, (model_obj, X_eval, name) in zip(axes, [
    (rf,  X_test,        'Random Forest'),
    (xgb, X_test,        'XGBoost'),
]):
    y_prob = model_obj.predict_proba(X_eval)
    for i, (cls, color) in enumerate(zip(le.classes_, roc_colors)):
        fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_prob[:, i])
        roc_auc_val = auc(fpr, tpr)
        ax.plot(fpr, tpr, color=color, lw=2, label=f'{cls} (AUC = {roc_auc_val:.3f})')
    ax.plot([0,1],[0,1], 'k--', lw=1)
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_title(f'ROC Curve — {name}', fontweight='bold')
    ax.legend(loc='lower right', fontsize=9)
    ax.set_xlim([0.0, 1.0])
    ax.set_ylim([0.0, 1.02])

plt.suptitle('RF and XGB ROC Curves for Credit Score Classification', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig("ROC_curve.png", dpi=300)
plt.show()
```

ROC curves measure the trade-off between True Positive Rate and False Positive Rate across all classification thresholds, providing a threshold-independent view of model performance. For this multi-class problem, a One-vs-Rest (OvR) strategy is applied, generating one ROC curve per class. AUC (Area Under the Curve) values approaching 1.0 indicate near-perfect discriminative ability. The `Good` class — the smallest and most imbalanced — achieving high AUC confirms that SMOTE successfully improved the model's ability to identify creditworthy individuals without artificially inflating performance.

## 22. Future Work

1. **Integration of real telecom and mobile money data.** Integrating mobile money transaction patterns, airtime usage, and digital wallet activity would significantly enrich the feature space and align the model more closely with the financial inclusion use case in markets such as Nigeria.

2. **Temporal modelling.** The dataset contains monthly snapshots per customer. A future iteration could exploit this longitudinal structure using recurrent neural networks or time-series feature engineering to capture trends in creditworthiness over time.

3. **Explainability with SHAP.** Deploying SHAP (SHapley Additive exPlanations) would enable per-customer explanations, telling an individual applicant exactly which factors influenced their score and by how much — a regulatory requirement in many jurisdictions.

4. **Fairness auditing.** A formal fairness assessment across demographic groups (age, occupation) using metrics such as equalised odds and demographic parity should be conducted before any production deployment.

5. **Hyperparameter optimisation.** Bayesian search or `Optuna`-based tuning across a wider hyperparameter space for XGBoost and Random Forest could yield further performance improvements, particularly in recall for the `Good` minority class.
